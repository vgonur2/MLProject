# Section 2: Exploratory Data Analysis

Certain types of descriptive statistics can help give us a better understanding of my dataset. To begin, some types of measures of center like the mean, can help give us a better understanding of the average nuber of points/rebounds/steals/games played etc. Using this information we can compare it to other teams in the NBA to see if the Los Angeles Lakers where a good team on a plyer to player basis. Furthermore certain measures of spread like finding the standard deviation of points/rebounds/steals/games played etc, can help us find out if the average of each players in those respective statistics where varied or not. 

Below is some of the code that finds the standard devaition and the mean of every single statistic. As we can see, some of the main things we noticed is that the mean games played per game is only 41 games even though there are 82 games in a season, however the standard devation for this is 28 games. The reason the mean is so low, and the standard deviaition is high is becuase the lakers had 21 people on their roster but can only put 5 guys out on the court at a time. Due to this, the starters like, Lebron, Anthony Davis, Austin Reaves, D'angelo Rusell, and Rui Hachimura, all get preference in games played, while guys at the end of the list, often times ride the bench and dont get much playing time. In a game the Lakers also only play about 8 - 10 of the 21 guys, and can only dress 15 of the 21 players on their roster for a game. This once again contributes to the low mean and the high standard deviation. This same phenomeon occurs with minutes per game (mean: 16, std: 27), and the reason is the exact same as above, there can be only 5 people out on the court, and the lakers have 21 people on their roster. One thing we can see is the mean and standard deviation for things such as steals per game and blocks per game, were both below 1spg/1bpg. This is becuase there are not many blocks and steals that happen in a game, and due to this the player averanges are going to be low. One thing that can be taken away from using descriptive statistics, is that a players time on the court greatly affects their stats, the longer a plyer is on a court on average they have higher stats per category.

```python
meanGP = df["Game Played"].mean()
stdGP = df["Game Played"].std()

meanGS = df["Games Started"].mean()
stdGS = df["Games Started"].std()

meanMPG = df["Minutes Per Game"].mean()
stdMPG = df["Minutes Per Game"].std()

meanPPG = df["Points Per Game"].mean()
stdPPG = df["Points Per Game"].std()

meanRebounds = df["Rebounds"].mean()
stdRebounds = df["Rebounds"].std()

meanAPG = df["Assists Per Game"].mean()
stdAPG = df["Assists Per Game"].std()

meanSPG = df["Steals Per Game"].mean()
stdSPG = df["Steals Per Game"].std()

meanBPG = df["Blocks Per Game"].mean()
stdBPG = df["Blocks Per Game"].std()

print("Mean Games Played ->" , meanGP , " | Std Games Played ->" , stdGP)
print("Mean Games Started ->" , meanGS , " | Std Games Started ->" , stdGS)
print("Mean Minutes Per Game ->" , meanMPG , " | Std Minutes Per Game ->" , stdMPG)
print("Mean Points Per Game ->" , meanPPG , " | Std Points Per Game ->" , stdPPG)
print("Mean Rebounds ->" , meanRebounds , " | Std Rebounds ->" , stdRebounds)
print("Mean Assists Per Game ->" , meanAPG , " | Std Assists Per Game ->" , stdAPG)
print("Mean Steals Per Game ->" , meanSPG , " | Std Steals Per Game ->" , stdSPG)
print("Mean Blocks Per Game ->" , meanBPG , " | Std Blocks Per Game ->" , stdBPG)






```

# Section 5 Data Science

The question that I am choosing to answer is to find the correlation between a players points per game compared to the amount of games that player has started in. I think that if a person has a higher number points in a game. The reason I believe this is becuase many times when you start the game, you want to have a big enough lead early on, so that the preassure of coming back from a deficit is not mounting on your head. Due to this you would usually want your best scorers to the be the ones starting. I want to make a machine learning to model to figure out the correlation between these two piece of data, create a regression line, compare these two piece of data within the existing players, and find out a players predicted number of games started for the season, dependent on the number of points they scored for that season. 

The way I plan on using python to solve this is by first importing LinearRegression from sklearn.linear_model. After this I am going to start by making linear regression models. So I make my independent variables "Points per game" and my dependent variables "Minutes Per game", then define my model, and do model.fit for my independent and dependent variables. Then, I will make a new coloumn in the dataframe titled, "Predicted Games Started". Lastly I will define that coloumn to be the models prediction based upon the independent variables. From here we can compare the models predicted minutes per game vs the actual minutes per game that players recieves depending on his statistics.

From the code below we can notice that the R value between Games Started and Points Per Game is about 0.94


```python
from sklearn.linear_model import LinearRegression
import pandas as pd
df1 = pd.read_csv("Lakers stats 2023 -2024 - Sheet1 (1).csv")



ind = ["Points Per Game"]
dep = "Games Started"



model = LinearRegression()
model.fit(df1[ind], df1[dep])



df1["Predicted Games Started"] = model.predict(df1[ind])


# To finish the code, we'll display the DataFrame:
df1

```
